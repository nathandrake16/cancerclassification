```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import joblib
from sklearn.decomposition import PCA
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

###Detect ID column and align gene matrices (keep only common genes)

```python
#detect ID column and align gene matrices (keep only common genes)
def detect_id_column(df, preferred=None):
    if preferred is None:
        preferred = ["sample_id","SampleID","patient_id","PATIENT_ID",
                     "case_id","bcr_patient_barcode","id","ID"]
    for name in preferred:
        if name in df.columns:
            return name
    obj_cols = df.select_dtypes(include=["object","category"]).columns.tolist()
    for c in obj_cols:
        if df[c].nunique() >= int(0.9*len(df)) and df[c].nunique() > 5:
            return c
    return df.columns[0]

def align_common_gene_columns(gdfs):
    """
    Input: list of gene-expression dataframes (one per cancer type).
    Output: aligned gene dataframes (each keeps only the intersection of gene cols),
            canonical ID name ("SAMPLE_ID"), list of common gene column names.
    """
    id_cols = [detect_id_column(g) for g in gdfs]
    gene_sets = []
    for g, idc in zip(gdfs, id_cols):
        gene_cols = [c for c in g.columns if c != idc]
        gene_sets.append(set(gene_cols))
    common_genes = sorted(list(set.intersection(*gene_sets)))
    aligned = []
    for g, idc in zip(gdfs, id_cols):
        sub = g[[idc] + common_genes].copy()
        sub = sub.rename(columns={idc: "SAMPLE_ID"})
        aligned.append(sub)
    return aligned, "SAMPLE_ID", common_genes

# NOTE: expects `gene_dataframes` (list) and `cancer_types` variables already defined in your notebook
```

###Align genes, concatenate, standardize, PCA

```python
#align gene matrices, concatenate, scale, global PCA
# PARAMETERS (change if you want fixed #components instead of variance-based)
PCA_VARIANCE = 0.95     # set float (e.g., 0.95) to keep that fraction of explained variance
PCA_N_COMPONENTS = None # or set integer, e.g., 100, to use fixed number instead of variance


print("[STEP] Aligning gene matrices...")
gene_aligned, gene_id_col, common_gene_cols = align_common_gene_columns(gene_dataframes)
print(f"[INFO] Common gene count: {len(common_gene_cols)}")

# concat and label
gene_blocks = []
for ct, gdf in zip(cancer_types, gene_aligned):
    tmp = gdf.copy()
    tmp["cancer_type"] = ct
    gene_blocks.append(tmp)
gene_all = pd.concat(gene_blocks, axis=0, ignore_index=True)
print("[INFO] Total samples (genes):", gene_all.shape[0])

# numeric matrix
X_genes = gene_all[common_gene_cols].astype(float).values
y_genes = gene_all["cancer_type"].values
ids_genes = gene_all[gene_id_col].values

# scale then PCA
print("[STEP] Scaling gene matrix and running PCA (global)...")
scaler_g = StandardScaler(with_mean=True, with_std=True)
Xg_scaled = scaler_g.fit_transform(X_genes)

if PCA_N_COMPONENTS is None:
    pca = PCA(n_components=PCA_VARIANCE, svd_solver="full", random_state=42)
else:
    pca = PCA(n_components=PCA_N_COMPONENTS, random_state=42)

Xg_pca = pca.fit_transform(Xg_scaled)
print(f"[INFO] PCA components: {Xg_pca.shape[1]}  Expl.Var.Sum: {pca.explained_variance_ratio_.sum():.4f}")

# build gene PCA DF
pca_cols = [f"GX_PC{i+1}" for i in range(Xg_pca.shape[1])]
gene_pca_df = pd.DataFrame(Xg_pca, columns=pca_cols)
gene_pca_df[gene_id_col] = ids_genes
gene_pca_df["cancer_type"] = y_genes
print(Xg_pca.shape[1])

# NOTE: expects `gene_dataframes` (list) and `cancer_types` variables already defined in your notebook
```

###Combine clinical dataframes into one

```python
# combine clinical dataframes (they already were trimmed to common clinical columns earlier)
print("[STEP] Combining clinical dataframes...")
clin_id_guess = detect_id_column(clinical_dataframes[0])

clin_blocks = []
for ct, cdf in zip(cancer_types, clinical_dataframes):
    c = cdf.copy()
    local_id = clin_id_guess if clin_id_guess in c.columns else detect_id_column(c)
    c = c.rename(columns={local_id: gene_id_col})  # rename clinical ID to SAMPLE_ID
    c["cancer_type"] = ct
    clin_blocks.append(c)

clinical_all = pd.concat(clin_blocks, axis=0, ignore_index=True)
print("[INFO] Total clinical rows:", clinical_all.shape[0])

# Quick class distribution check (per cancer)
print("\nSample counts per cancer (gene side) and clinical side:")
print(gene_all["cancer_type"].value_counts().sort_index())
print(clinical_all["cancer_type"].value_counts().sort_index())
```

###Merge clinical & PCA gene features into master dataset

```python
#  merge clinical + gene PCA features (inner join on SAMPLE_ID + cancer_type)
print("[STEP] Merging clinical and gene PCA features (inner join)..")
master = pd.merge(clinical_all, gene_pca_df, on=[gene_id_col, "cancer_type"], how="inner")
print(master['cancer_type'].value_counts())
print("[INFO] Master shape (rows, cols):", master.shape)
print("[INFO] Columns preview:", list(master.columns[:20]))
display(master.head(3))
```

###Clean clinical columns (numeric coercion + categorical cast)

```python
# cleaning clinical features before preprocessing
# Identify feature groups
drop_cols = [gene_id_col, "cancer_type"]
feature_cols = [c for c in master.columns if c not in drop_cols]
pca_feature_cols = [c for c in feature_cols if c.startswith("GX_PC")]
clinical_feature_cols = [c for c in feature_cols if c not in pca_feature_cols]

# Detect categorical vs numeric using dtypes
clin_cat = master[clinical_feature_cols].select_dtypes(include=["object","category"]).columns.tolist()
clin_num = [c for c in clinical_feature_cols if c not in clin_cat]

# Coerce clinical numeric columns to numeric (safe) and impute missing with median
for col in clin_num:
    master[col] = pd.to_numeric(master[col], errors="coerce")
    median = master[col].median()
    master[col] = master[col].fillna(median)

# Force categorical columns to strings and fill missing with "missing"
for col in clin_cat:
    master[col] = master[col].astype(str).fillna("missing")

print("[INFO] Clinical numeric:", clin_num)
print("[INFO] Clinical categorical:", clin_cat)
print("[INFO] PCA feature count:", len(pca_feature_cols))
```

###Preprocessing pipeline (ColumnTransformer)

```python
# build preprocessor (scaling numeric, one-hot categorical, passthrough gene PCs)
# Use compatibility guard for OneHotEncoder arg name
try:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
except TypeError:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)

transformers = []
if clin_num:
    transformers.append(("clin_num", StandardScaler(), clin_num))
if clin_cat:
    transformers.append(("clin_cat", ohe, clin_cat))
if pca_feature_cols:
    transformers.append(("genes", "passthrough", pca_feature_cols))

preprocessor = ColumnTransformer(transformers=transformers, remainder="drop")
# Build X, y for modeling
X = master[clin_num + clin_cat + pca_feature_cols]
y = master["cancer_type"].values

print("[INFO] Preprocessor created. X shape:", X.shape)
```

###Stratified train/test split

```python
# stratified train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
print(f"[INFO] Train samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
```

###Baseline 1: Multinomial Logistic Regression (CV + test eval)

```python
# Logistic Regression baseline (stratified CV on train + evaluation on test)
lr_pipeline = Pipeline([
    ("prep", preprocessor),
    ("clf", LogisticRegression(multi_class="multinomial", solver="saga",
                               max_iter=5000, n_jobs=-1, random_state=42))
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
print("[MODEL] Logistic Regression: running CV on training set...")
cv_scores = cross_val_score(lr_pipeline, X_train, y_train, cv=cv, scoring="accuracy", n_jobs=-1)
print(f"[CV] LR Accuracy: mean={cv_scores.mean():.4f} Â± {cv_scores.std():.4f}")

# Fit on train and evaluate on test
lr_pipeline.fit(X_train, y_train)
y_pred = lr_pipeline.predict(X_test)

print("\n[Test] LR classification report:")
print(classification_report(y_test, y_pred, digits=4))

cm = confusion_matrix(y_test, y_pred, labels=sorted(np.unique(y)))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=sorted(np.unique(y)))
plt.figure(figsize=(7,6))
disp.plot(xticks_rotation=45)
plt.title("LR Confusion Matrix")
plt.tight_layout()
plt.show()
```

###Baseline 2: Random Forest (CV + test eval)

```python
# Random Forest baseline
rf_pipeline = Pipeline([
    ("prep", preprocessor),
    ("rf", RandomForestClassifier(n_estimators=400, class_weight="balanced_subsample",
                                  random_state=42, n_jobs=-1))
])

print("[MODEL] Random Forest: running CV on training set...")
cv_scores_rf = cross_val_score(rf_pipeline, X_train, y_train, cv=cv, scoring="accuracy", n_jobs=-1)
print(f"[CV] RF Accuracy: mean={cv_scores_rf.mean():.4f} Â± {cv_scores_rf.std():.4f}")

rf_pipeline.fit(X_train, y_train)
y_pred_rf = rf_pipeline.predict(X_test)

print("\n[Test] RF classification report:")
print(classification_report(y_test, y_pred_rf, digits=4))

cm2 = confusion_matrix(y_test, y_pred_rf, labels=sorted(np.unique(y)))
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=sorted(np.unique(y)))
plt.figure(figsize=(7,6))
disp2.plot(xticks_rotation=45)
plt.title("RF Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
import numpy as np

# One-hot encode the target variable
y_train_encoded = to_categorical(np.array([np.where(np.unique(y_train) == c)[0][0] for c in y_train]))
y_test_encoded = to_categorical(np.array([np.where(np.unique(y_test) == c)[0][0] for c in y_test]))

# Apply the preprocessor to the training and testing data
X_train_processed = preprocessor.transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Reshape the data for CNN (add a channel dimension)
X_train_cnn = X_train_processed.reshape(X_train_processed.shape[0], X_train_processed.shape[1], 1)
X_test_cnn = X_test_processed.reshape(X_test_processed.shape[0], X_test_processed.shape[1], 1)

# Build the CNN model
model_cnn = Sequential()
model_cnn.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train_cnn.shape[1], 1)))
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(Dropout(0.3))
model_cnn.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(Dropout(0.3))
model_cnn.add(Flatten())
model_cnn.add(Dense(64, activation='relu'))
model_cnn.add(Dropout(0.3))
model_cnn.add(Dense(len(np.unique(y)), activation='softmax'))

# Compile the model
model_cnn.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Define early stopping
early_stopping_cnn = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history_cnn = model_cnn.fit(X_train_cnn, y_train_encoded,
                            epochs=100,
                            batch_size=32,
                            validation_split=0.2,
                            callbacks=[early_stopping_cnn],
                            verbose=1)

# Evaluate the model on the test set
loss_cnn, accuracy_cnn = model_cnn.evaluate(X_test_cnn, y_test_encoded, verbose=0)
print(f"Test Loss (CNN): {loss_cnn:.4f}")
print(f"Test Accuracy (CNN): {accuracy_cnn:.4f}")

# Plot training history (optional)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_cnn.history['accuracy'])
plt.plot(history_cnn.history['val_accuracy'])
plt.title('CNN Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(history_cnn.history['loss'])
plt.plot(history_cnn.history['val_loss'])
plt.title('CNN Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
